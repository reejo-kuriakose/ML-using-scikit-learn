```python
x = df.iloc[:,[3,4]].values
```

```python
import numpy as np
np.sqrt(200)
```

the no of clusters k is chosen from 1 to sqrt of the total number of values (1 to 14)

best value of k is supposed to be between 2 and 14

```python
# ELBOW METHOD
k = range(2,14)
sse=[]

from sklearn.cluster import KMeans
for i in k:
  model_demo = KMeans(n_clusters = i, random_state = 0)
  model_demo.fit(x)
  sse.append(model_demo.inertia_)
plt.scatter(k,sse)
plt.plot(k,sse)
plt.show()
```

```python
# k is considered for this dataset
k = 5
from sklearn.cluster import KMeans
model = KMeans(n_clusters=k,random_state=0)
model.fit(x)
y = model.predict(x)
y
```

```python
np.unique(y,return_counts=True)
# cluster 0 35 values
# cluster 1 81 values etc
```

```python
model.cluster_centers_
#centroids of all the 5 clusters
```

```python
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1])
```

```python
for i in range(k): 
  plt.scatter(x[y==i,0],x[y==i,1],label = f"Cluster {i}")
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1])
plt.legend()
plt.show()
```
