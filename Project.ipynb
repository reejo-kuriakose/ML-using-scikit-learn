```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/arib168/data/main/50_Startups.csv")
df
```

```python
# replacing 0s with mean value
df.iloc[:,0]=df.iloc[:,0].replace(0,df.iloc[:,0].mean())
df.iloc[:,2]=df.iloc[:,2].replace(0,df.iloc[:,2].mean())
df
```

```python
import matplotlib.pyplot as plt
x = ["MIN", "MAX"]
y = [df['R&D Spend'].max(),df['R&D Spend'].min()]
plt.plot(x,y)
plt.ylabel('R&D Spend')
plt.show()
```

```python
import matplotlib.pyplot as plt
x = ["MIN", "MAX"]
y = [df['Administration'].max(),df['Administration'].min()]
plt.bar(x,y,color=["blue","red"])
plt.title('Administeration')
plt.ylabel('Administration')
plt.show()
```

```python
import matplotlib.pyplot as plt
label = ["MIN", "MAX"]
values = [df['Profit'].max(),df['Profit'].min()]
plt.pie(values, shadow=True, colors=["red","yellow"], autopct=lambda p:f'{p*sum(values)/100 :.0f}',startangle=180)
plt.title("Profit")
plt.legend(labels=label)
plt.show()
```

```python
# creating input and output data for the model
x = df.iloc[:,0:3]
y = df.iloc[:,4]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_pred,y_test)
```
