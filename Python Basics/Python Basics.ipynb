SEQUENCES IN PYTHON \
sequences: containers where multiple items can be stored in an order (a=(5,6,7,8,9)) \
variable: stores a single element (a=5) \
arrays: collection of elements of similar datatypes and stored in contagious  memory locations \
TYPES OF SEQUENCES \
String \
Lists \
Tuples \
Dictionary \
Sets

LISTS: most versatile sequence/datatype in python \
       stores elements of different datatype \
       elements are seperated by commas \
       represented in square brackets \
       allows repetition of elements \
       mutable(addition & deletion of elements is possible)
EXAMPLE: 
  a=[9,3,1,"abc",1,4]
  type(a)

```python
a=[1,2,3,4,5]
a[::-1] # syntax: list[start:stop:step] (reverse a list)
```

```python
# operations on a list
fruits = ["grape","orange","mango","strawberry","apple"]
fruits[3][0:5] # 0(inclusive) 5(exclusive)
```

```python
# add an element to a list(at the end) append() -
fruits.append("kiwi")
fruits
```

```python
# insert in between
fruits.insert(2,"apple")
fruits
```

```python
fruits.remove("kiwi")
fruits
```

```python
fruits.pop(2) # to delete using index
fruits
```

```python
fruits.reverse() # to reverse the list
fruits
```

```python
fruits*2
```

```python
# extend the list
veggies=["carrot","bell peppers", 'tomato','onion']
veggies
```

```python
# extend from another list
fruits.extend(veggies)
fruits
```

```python
# sorting in alphabetic order
fruits.sort()
fruits
```

```python
fruits.sort(reverse=True)
fruits
```

DICTIONARY \
{key:value,key:value,key:value} \
ordered collection of elements \
mutable(can add or remove key value pairs from a dictionary) \
no indexing, no slicing

```python
data = {1:'python',2:'java',3:'javascript',5:'c++'}
type(data)
```

```python
data[5] # retrive data
print(data.get(3,"Key not found"))
print(data.get(4,"Key not found")) # display the statement if value does not exist
```

```python
# generate a dictionary from multiple list
keys = ['person 1', 'person 2', 'person 3', 'person 4']
values = ['pizza', 'burger', 'fries', 'pasta']
data = dict(zip(keys,values)) # zip() combines both the lists and refers to them as key value pairs accordingly
data
```

```python
# to add a key:value pair to the dictionary
data['person 5'] = 'noodles'
data
```

```python
# to remove/delete a key:value pair
del data['person 3']
data
```

```python
data['person 1'] = ['pizza','burger']
data
```

```python
data['person 1'] = ['pizza', 'burger', 'lasgna']
data
```

```python
data['person 1'][2]
```

```python
import matplotlib.pyplot as plt
```

```python
a = [1,2,3,4,5,6,7]
b = [5,4,6,2,3,1,4]
c = [1,4,8,4,2,3,6]

plt.title("MATPLOTLIB LINE GRAPH",fontsize=15)
plt.plot(a,b,color='red',label='line 1', marker="o")
plt.plot(a,c,color='blue',label='line 2')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid()
plt.legend()
plt.show()
```

```python
a = [1,2,3,4,5,6,7]
b = [5,4,6,2,3,1,4]
c = [1,4,8,4,2,3,6]

plt.title("MATPLOTLIB LINE GRAPH",fontsize=15)
plt.plot(a,b,color='red',label='line 1', marker="o", markersize=5, linestyle = 'dotted', linewidth=5)
plt.plot(a,c,color='blue',label='line 2', marker="*", markeredgecolor='red', markerfacecolor='red')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(color='grey', linestyle="--", linewidth=2)
plt.legend()
plt.show()
```

```python
import numpy as np
x = np.arange(0,10,0.2)
y = np.sin(x)
plt.plot(x,y,color='red')
plt.fill_between(x,y,color='red',alpha=0.3) # alpha for transparency
plt.axvline(3.175,color='blue')
plt.axvline(6.25,color='green')
plt.axhline(0,color='black')
```

```python
# BAR GRAPH
name = ['AA', 'BB', 'CC', 'DD', 'EE']
weight = [50,42, 67, 33, 81]
ax = plt.axes()
ax.set_facecolor('grey')
plt.bar(name,weight,color='red')
```

```python
# BAR GRAPH
name = ['AA', 'BB', 'CC', 'DD', 'EE']
weight = [50, 42, 67, 33, 81]
col12 = ['red','yellow','green','blue','orange']
plt.bar(name,weight,color=col12)
```

```python
# HORIZONTAL BAR GRAPH
name = ['AA', 'BB', 'CC', 'DD', 'EE']
weight = [50, 42, 67, 33, 81]
col12 = ['red','yellow','green','blue','orange']
plt.barh(name,weight,color=col12)
plt.savefig("graph.png") # to save as png
```

```python
# PIE CHART

hours = [10,8,2,4]
work = ['college','sleep','food','leisure/mobile']
c = ['#46ff00','b','r','#00FBFF']
plt.pie(hours,labels=work,colors=c,explode=[0,0,0.3,0.2],labeldistance=1.1,shadow=True,startangle=90,counterclock=False, autopct='%1.2f%%') # explode to seperate the slices # default anticlockwise
plt.legend(loc='lower right')
plt.show()
```

```python
# SCATTER PLOT
lst1 = [1,2,3,4,5,6,7,8]
lst2 = [4,5,6,2,3,5,8,9]
plt.scatter(lst1,lst2,s=100,c='red',marker='^',alpha=0.3)
plt.show()
```
