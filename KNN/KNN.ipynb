```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/arib168/data/main/Underweight-knn.csv")
df
```

```python
x = df.iloc[:,[0,1]].values
y = df.iloc[:,-1].values
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=3, metric='euclidean') # root of 11 (no. of values) is 3.xx, since 4 is even taking 3 
```

```python
model.fit(x,y)
```

```python
y_pred = model.predict(x)
y_pred
```

```python
y
```

```python
model.predict([[55,165]])
```

```python
model.score(x,y)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y,y_pred)
```

```python
# MULTI CLASS CLASSIFICATION USING IRIS DATASET
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris
```

```python
iris.feature_names # print input columns
```

```python
iris.target_names # print output columns
```

```python
iris.target
```

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names) #create dataframe from this imported dataset of iris
df['Output'] = iris.target
df
```

```python
x = df.iloc[:,0:4].values
y = df.iloc[:,-1].values
```

```python
import numpy as np
np.unique(x,return_counts=True)
```

```python
np.unique(y,return_counts=True)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=13)
model.fit(x,y)
```

k can be 11 or 13 and check accuracy \
k = 13 gives better accuracy \
default metric is minowski with power 2 which is euclidian

```python
y_pred = model.predict(x)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y,y_pred)
```

```python
model.predict([[5.9,3.0,5.1,1.8]])
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y,y_pred)
import seaborn as sns
sns.heatmap(confusion_matrix(y,y_pred), annot=True,cmap='hot')
```

```python
%%writefile app.py
import streamlit as st
import numpy as np
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier

st.title("IRIS FLOWER CLASSIFICATION")

var = load_iris()

x = var.data
y = var.target
model = KNeighborsClassifier(n_neighbors = 13)

model.fit(x,y)

xmin = np.min(x, axis=0)
xmax = np.max(x, axis=0)

sepal_length = st.slider("Sepal Length",float(xmin[0]),float(xmax[0]))
sepal_width = st.slider("Sepal Width",float(xmin[1]),float(xmin[1]))
petal_length = st.slider("Pepal Length",float(xmin[2]),float(xmax[2]))
petal_width = st.slider("Petal Width",float(xmin[3]),float(xmax[3]))


y_pred = model.predict([[sepal_length,sepal_width,petal_length,petal_width]])

op = ['Iris-setosa','Iris-versicolor','Iris-virginica']
st.title(op[y_pred[0]])
```
