```python
# To impove the accuracy, we need to scale the data to get it within the required range of 0 to 1
import numpy as numpy
import matplotlib.pyplot as plt

np.random.seed(153)
x1 = np.random.randint(18,60,30)
x1 = np.sort(x1)
x2 = np.random.randint(15000,150000,30)
plt.plot(x1,x2)
```

```python
# MinMax Scaler
x1min = min(x1)
x2min = min(x2)

x1max = max(x1)
x2max = max(x2)

x1norm = (x1 - x1min)/(x1max - x1min)
x2norm = (x2 - x2min)/(x2max - x2min)

plt.plot(x1norm,x2norm)
```

```python
# normalization is only applied to input data
from sklearn.preprocessing import MinMaxScaler
scale = MinMaxScaler()
x_train = scale.fit_transform(x_train)
x_test = scale.transform(x_test)
```

When the dataset is imbalanced, accuracy alone is not the metric to consider,
we also use precision, recall and f1-score for checking whether the predictions
are good or not

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred )
```

  TRUE POSITIVE = 67 (correctly predicted as 1) \
   FALSE POSITIVE = 1 \
   FALSE NEGATIVE = 10 \
   TRUE NEGATIVE = 22 (correctly predicted as 0)
  [TP, FP],
  [FN, TN]

```python
accuracy = (67+22)/(67+1+10+22)
accuracy
```

```python
precision = (67)/(67+1)
precision
```

```python
recall = (67)/(67+10)
recall
```

```python
f1_score = 2*(precision*recall)/(precision + recall)
f1_score
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_pred, y_test))
```
