```python
# SPAM HAM DATASET 
# CREATE A ML MODEL FOR PREDICTING IF A MESSAGE WAS SPAM OR NOT (SVM and NaiveBayes)

import pandas as pd
df = pd.read_table("https://raw.githubusercontent.com/arib168/data/main/spam.tsv")
df
```

```python
df.info()
```

```python
x = df['message'].values
y = df['label'].values

print(df['message'][896])
print(df['label'][896])
```

```python
df['label'].value_counts()
```

```python
# train test and split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state = 0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
# Applying the Count Vectorizer
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer(stop_words='english')
x_train_vect = vect.fit_transform(x_train)
x_test_vect = vect.transform(x_test)
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train_vect,y_train)
y_pred = model.predict(x_test_vect)
accuracy_score(y_pred,y_test)
```

```python
# Applying the ML algorithm to fit and predict
# SVM
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(x_train_vect, y_train)
```

```python
y_pred = model.predict(x_test_vect)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
# PIPELINE (combines two estimators = svc + countvecotizer)
# we use pipeline so that we need not vectorize the text every time we want to predict the output
from sklearn.pipeline import make_pipeline
model2 = make_pipeline(CountVectorizer(),SVC())

model2.fit(x_train,y_train)

y_pred2 = model2.predict(x_test)
y_pred2
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred2,y_test)
```

```python
# APPLY ML MODEL ( NAIVE BAYES )
from sklearn.naive_bayes import MultinomialNB
model3 = MultinomialNB()
```

```python
model3.fit(x_train_vect,y_train)
```

```python
y_pred3 = model3.predict(x_test_vect)
y_pred3
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred3, y_test)
```

```python
# PIPELINE (combines two estimators = MultinomialNM + countvecotizer)
# we use pipeline so that we need not vectorize the text every time we want to predict the output
from sklearn.pipeline import make_pipeline
model4 = make_pipeline(CountVectorizer(),MultinomialNB())

model4.fit(x_train,y_train)

y_pred4 = model4.predict(x_test)
y_pred4
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred4, y_test)
```
