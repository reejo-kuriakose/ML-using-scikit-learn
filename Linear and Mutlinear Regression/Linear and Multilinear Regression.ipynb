Linear Regression

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/Salary_Data.csv')
df
```

```python
df.info()
```

sklearn accepts x (input data) in 2 dimensions only \
x always needs to be in 2 dimensional Numpy array ONLY \
y can be 1 dimension

```python
x = df[['YearsExperience']].values # for numpy array
# OR
# x=df.iloc[:,[0]].values
x
```

```python
y = df['Salary'].values
# OR
# y = df.iloc[:,1].values
y
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y) # original data points
```

```python
from sklearn.linear_model import LinearRegression # import the algorithm
```

```python
model = LinearRegression() # calling the algorithm
```

```python
# training the machine/fitting the model
model.fit(x,y)

LinearRegression()
```

```python
# testing part of the model
y_pred = model.predict(x)
y_pred
```

```python
y
```

```python
import matplotlib.pyplot as plt
plt.scatter (x,y, label='original datap points')
plt.plot(x,y_pred,c='r',label = 'BEST FIT LINE')
plt.legend()
plt.show()
```

```python
# create a new dataframe for the actual vs predicted output
df_new = pd.DataFrame({'Actual Output':y,'Predicted Output':y_pred})
df_new
```

```python
# Accuracy (r2 score for regression)
from sklearn.metrics import r2_score
r2_score(y,y_pred)
```

```python
model.predict([[6.0]])
```

```python
model.coef_ # m value or slope
```

```python
model.intercept_ # c value or y-intercept
```

pickling - saving and loading objects using pickle module \
to save a model in scikit-learn by using Python's built-in persistence model \
serialization and deserialization

```python
import pickle
pickle.dump(model, open('linear_reg.pkl','wb')) # creates a pickle file for the linear regression model
```

```python
# deserialization step (reloading the model and using it directly for predictions)
import pickle
reload_model = pickle.load(open('linear_reg.pkl','rb'))
```

```python
reload_model.predict([[10.0]])
```

Multilinear Regression

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/CarPrice_Assignment.csv')
df
```

```python
df.info()
```

```python
cars_numeric = df.select_dtypes(include=['int64','float64'])
cars_numeric
```

```python
df.info()
```

```python
x = cars_numeric.iloc[:,0:13].values
y = cars_numeric.iloc[:,13].values
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
```

```python
x_test
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```
